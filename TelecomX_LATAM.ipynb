## üîπ Introducci√≥n
El objetivo de este an√°lisis es comprender el problema de **evasi√≥n de clientes (Churn)** en la empresa Telecom X.  
La evasi√≥n ocurre cuando los clientes deciden cancelar el servicio, lo que impacta directamente en los ingresos y sostenibilidad del negocio.  
Mediante el an√°lisis de datos demogr√°ficos, tipo de contrato, m√©todos de pago y comportamiento de consumo, buscamos identificar patrones que expliquen por qu√© algunos clientes permanecen y otros se retiran.

---

## üîπ Limpieza y Tratamiento de Datos
1. **Importaci√≥n de datos**: Se carg√≥ el archivo JSON desde la API de Telecom X.  
2. **Normalizaci√≥n**: Se expandieron las columnas anidadas (`customer`, `phone`, `internet`, `account`) para obtener un DataFrame plano.  
3. **Correcci√≥n de inconsistencias**:  
   - Eliminaci√≥n de duplicados.  
   - Conversi√≥n de columnas num√©ricas (`Charges.Monthly`, `Charges.Total`) a valores num√©ricos.  
4. **Estandarizaci√≥n y transformaci√≥n**:  
   - Conversi√≥n de variables binarias (`Yes/No`) a valores 1 y 0.  
   - Renombrado de columnas para mayor claridad (`MonthlyCharges`, `TotalCharges`).  
5. **Creaci√≥n de columna adicional**: Se gener√≥ `daily_accounts` para calcular el gasto diario a partir de la facturaci√≥n mensual.

---

## üîπ An√°lisis Exploratorio de Datos
- **Distribuci√≥n general de Churn**: Aproximadamente un cuarto de los clientes cancelaron el servicio.  
- **An√°lisis descriptivo**:  
  - Clientes con menor tiempo de contrato (`tenure`) presentan mayor evasi√≥n.  
  - Cargos mensuales altos (`MonthlyCharges`) se asocian con mayor probabilidad de cancelaci√≥n.  
  - Clientes con acumulados bajos (`TotalCharges`) tienden a cancelar m√°s, reflejando que son clientes nuevos.  
- **Variables categ√≥ricas**:  
  - El g√©nero no mostr√≥ diferencias significativas en evasi√≥n.  
  - Los contratos mensuales tienen mayor tasa de cancelaci√≥n que los contratos de largo plazo.  
  - Los m√©todos de pago electr√≥nicos presentan mayor evasi√≥n que pagos autom√°ticos o transferencias.  
- **Variables num√©ricas**:  
  - Histogramas mostraron que clientes con `tenure` bajo y `MonthlyCharges` altos concentran la mayor evasi√≥n.  

---

## üîπ Conclusiones e Insights
- La **evasi√≥n est√° fuertemente asociada con clientes nuevos** (tenure bajo).  
- Los **cargos mensuales altos** incrementan la probabilidad de cancelaci√≥n.  
- Los **contratos mensuales y m√©todos de pago electr√≥nicos** son los perfiles con mayor riesgo de churn.  
- El g√©nero no influye significativamente en la decisi√≥n de cancelar.  

---

## üîπ Recomendaciones
1. **Fidelizaci√≥n temprana**: Implementar estrategias de retenci√≥n en los primeros meses de contrato.  
2. **Planes m√°s atractivos**: Ofrecer contratos a largo plazo con beneficios claros.  
3. **Optimizaci√≥n de precios**: Revisar la estructura de cargos mensuales para clientes con alto riesgo.  
4. **Incentivos en m√©todos de pago**: Promover pagos autom√°ticos con beneficios adicionales.  
5. **Monitoreo continuo**: Crear dashboards para identificar clientes en riesgo y actuar de manera preventiva.

La informaci√≥n obtenida permitir√° dise√±ar acciones concretas para **reducir la tasa de churn y mejorar la fidelizaci√≥n de clientes**.


```python
import pandas as pd
import requests

# URL raw del archivo JSON
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Obtener datos desde la API (GitHub raw)
response = requests.get(url)
data = response.json()

# Convertir a DataFrame base
df = pd.DataFrame(data)

# Ver primeras filas
df.head()
```

```python
# Normalizar columnas anidadas
customer_df = pd.json_normalize(df['customer'])
phone_df    = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df  = pd.json_normalize(df['account'])

# Unir todo en un solo DataFrame
df_final = pd.concat([df[['customerID','Churn']], customer_df, phone_df, internet_df, account_df], axis=1)

# Explorar columnas y tipos de datos
print(df_final.dtypes)
print(df_final.columns)
```

```python
# Valores nulos
print(df_final.isnull().sum())

# Duplicados
print(df_final.duplicated().sum())

# Eliminar duplicados
df_final = df_final.drop_duplicates()

# Convertir columnas num√©ricas a tipo correcto
df_final['Charges.Monthly'] = pd.to_numeric(df_final['Charges.Monthly'], errors='coerce')
df_final['Charges.Total']   = pd.to_numeric(df_final['Charges.Total'], errors='coerce')
```

```python
# Mapear valores binarios
df_final['PaperlessBilling'] = df_final['PaperlessBilling'].map({'Yes':1, 'No':0})
df_final['StreamingTV'] = df_final['StreamingTV'].map({'Yes':1, 'No':0, 'No internet service':0})
df_final['StreamingMovies'] = df_final['StreamingMovies'].map({'Yes':1, 'No':0, 'No internet service':0})

# Renombrar columnas
df_final.rename(columns={'Charges.Monthly':'MonthlyCharges',
                         'Charges.Total':'TotalCharges'}, inplace=True)
```

```python
# Crear columna de cuentas diarias
df_final['daily_accounts'] = df_final['MonthlyCharges'] / 30

# Verificar primeras filas
df_final[['customerID','MonthlyCharges','daily_accounts']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Conteo general de Churn
plt.figure(figsize=(6,4))
sns.countplot(data=df_final, x='Churn')
plt.title("Distribuci√≥n general de Churn")
plt.show()
```

```python
# Resumen estad√≠stico general
print(df_final.describe())

# M√©tricas espec√≠ficas
print("Tenure (meses):")
print("Media:", df_final['tenure'].mean())
print("Mediana:", df_final['tenure'].median())
print("Desviaci√≥n est√°ndar:", df_final['tenure'].std())

print("\nMonthlyCharges:")
print("Media:", df_final['MonthlyCharges'].mean())
print("Mediana:", df_final['MonthlyCharges'].median())
print("Desviaci√≥n est√°ndar:", df_final['MonthlyCharges'].std())

print("\nTotalCharges:")
print("Media:", df_final['TotalCharges'].mean())
print("Mediana:", df_final['TotalCharges'].median())
print("Desviaci√≥n est√°ndar:", df_final['TotalCharges'].std())
```

```python
# Distribuci√≥n de Churn por g√©nero
print(df_final.groupby('gender')['Churn'].value_counts(normalize=True))

# Distribuci√≥n de Churn por tipo de contrato
print(df_final.groupby('Contract')['Churn'].value_counts(normalize=True))

# Distribuci√≥n de Churn por m√©todo de pago
print(df_final.groupby('PaymentMethod')['Churn'].value_counts(normalize=True))
```

```python
# Distribuci√≥n de Tenure por Churn
plt.figure(figsize=(8,5))
sns.histplot(data=df_final, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title("Distribuci√≥n de Tenure (meses) seg√∫n Churn")
plt.show()

# Distribuci√≥n de MonthlyCharges por Churn
plt.figure(figsize=(8,5))
sns.histplot(data=df_final, x='MonthlyCharges', hue='Churn', multiple='stack', bins=30)
plt.title("Distribuci√≥n de MonthlyCharges seg√∫n Churn")
plt.show()

# Distribuci√≥n de TotalCharges por Churn
plt.figure(figsize=(8,5))
sns.histplot(data=df_final, x='TotalCharges', hue='Churn', multiple='stack', bins=30)
plt.title("Distribuci√≥n de TotalCharges seg√∫n Churn")
plt.show()
```

```python
# 1. Calcular matriz de correlaci√≥n
corr_matrix = df_final[['daily_accounts','MonthlyCharges','TotalCharges','tenure']].corr()

# 2. Mostrar matriz de correlaci√≥n como mapa de calor
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correlaci√≥n entre variables num√©ricas")
plt.show()

# 3. Relaci√≥n entre daily_accounts y Churn
plt.figure(figsize=(8,6))
sns.boxplot(data=df_final, x='Churn', y='daily_accounts')
plt.title("Relaci√≥n entre cuentas diarias y Churn")
plt.show()

# 4. Relaci√≥n entre cantidad de servicios contratados y Churn
# Crear columna con n√∫mero de servicios contratados
df_final['num_services'] = (
    (df_final[['PhoneService','MultipleLines','InternetService','OnlineSecurity',
               'OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
               'StreamingMovies']] == 'Yes').sum(axis=1)
)

plt.figure(figsize=(8,6))
sns.boxplot(data=df_final, x='Churn', y='num_services')
plt.title("Relaci√≥n entre n√∫mero de servicios contratados y Churn")
plt.show()
```
